```python
def compute_mean_std(loader):
    channels_sum, channels_squared_sum, num_batches = 0, 0, 0

    for data, _ in loader:
        # print(data.size())
        channels_sum += torch.mean(data, dim=[0, 2, 3])
        channels_squared_sum += torch.mean(data ** 2, dim=[0, 2, 3])
        num_batches += 1

    mean = channels_sum / num_batches
    std = (channels_squared_sum / num_batches - mean ** 2) ** 0.5
    print(mean, std)
    return mean, std
```

```python
import os
import shutil

# Define the paths to your 'train', 'test', and 'valid' folders

base_path = "/home/jamontanac/Documents/AdversarialExamples/adversarialexamples/data/01_raw/art_dataset"
train_path = os.path.join(base_path, 'train')
valid_path = os.path.join(base_path, 'valid')

# List all subdirectories in both 'train' and 'valid'
train_subdirs = [d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))]
valid_subdirs = [d for d in os.listdir(valid_path) if os.path.isdir(os.path.join(valid_path, d))]

# Identify matching subdirectories
matching_subdirs = set(train_subdirs).intersection(valid_subdirs)

```

```python
for subdir in matching_subdirs:
    valid_subdir_path = os.path.join(valid_path, subdir)
    train_subdir_path = os.path.join(train_path, subdir)
    
    for file in os.listdir(valid_subdir_path):
        src_file_path = os.path.join(valid_subdir_path, file)
        dst_file_path = os.path.join(train_subdir_path, file)
        
        # Move the file
        shutil.move(src_file_path, dst_file_path)
        print(f"Moved {file} from {valid_subdir_path} to {train_subdir_path}")

print("All matching files have been moved.")
```
