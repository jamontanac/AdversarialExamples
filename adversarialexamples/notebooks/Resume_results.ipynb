```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
models = ["Regnet_y", "Regnet_x", "Resnet"]
attacks = ["CarliniL2", "PGD", "DeepFool", "FSGM"]
results = []  # List to hold all results for easy plotting

for attack in attacks:
    for model in models:
        report = catalog.load(f'{model}_model_Report_{attack}@Dataset')
        for defense, data in report.items():
            accuracies = data['Accuracies']
            # Append a dictionary for each type of accuracy
            results.append({'Model': model, 'Defense': defense, 'Type': 'Original', 'Accuracy': accuracies['Original']})
            results.append({'Model': model, 'Defense': defense, 'Type': 'Adversarial', 'Accuracy': accuracies['Adversarial']})
            results.append({'Model': model, 'Defense': defense, 'Type': 'Defense', 'Accuracy': accuracies['Defense']})
```

```python
df=pd.DataFrame(results)
df.query("Type == 'Defense'").groupby(['Model','Defense'])['Accuracy'].agg(['mean','std','max','min']).sort_values(by=['mean','max'],ascending=False)
```

```python
import math
# Convert results to a DataFrame for easy plotting
df = pd.DataFrame(results)
df = df.query("Type != 'Original'")
df = df[df['Defense'].str.contains("JPEG")]
unique_defenses = df.Defense.unique()
# Assuming unique_defenses is already defined
num_defenses = len(unique_defenses)
num_columns = 3
num_rows = math.ceil(num_defenses / num_columns)

# Initialize the figure with a specific size
fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, num_rows * 6), constrained_layout=True)

# Flatten the axes array for easy iteration if there's more than one row
if num_rows > 1:
    axes = axes.flatten()
else:
    axes = [axes]

# Loop over each unique defense and create its plot
for i, defense in enumerate(unique_defenses):
    ax = axes[i]
    df_defense = df[df['Defense'] == defense]

    # Use seaborn to create the box plot for the current defense on the specified axes
    sns.boxplot(x='Model', y='Accuracy', hue='Type', data=df_defense, ax=ax)

    # Set titles and labels
    ax.set_title(f'Defense: {defense}')
    ax.set_ylabel('Accuracy (%)')
    ax.set_xlabel('Model')


# Hide any unused axes if the number of plots is not a perfect multiple of the layout
for i in range(len(unique_defenses), num_columns * num_rows):
    fig.delaxes(axes[i])

# Automatically adjust subplot params so that the subplot(s) fits in to the figure area
plt.tight_layout()

# Show the plot
plt.show()
```
